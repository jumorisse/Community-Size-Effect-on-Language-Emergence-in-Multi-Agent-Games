```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager

plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

```python
comm_sizes = [1,2,4]

sizes = []
accs = []
avg_accs = []

with open('/home/jmorisse/Desktop/BA Thesis/Bachelor-Thesis/Code/Data/Results/accuracies.txt', 'r') as f:
    for i,line in enumerate(f):
        seed_accs = [float(acc) for acc in line.split(';')[0].split(',')]
        for j in range(len(seed_accs)):
            sizes.append(str(comm_sizes[i]))
        seed_avg = float(line.split(';')[1])
        
        for seed_acc in seed_accs:
            accs.append(seed_acc)
        avg_accs.append(seed_avg)

print(sizes)
print(accs)
print(avg_accs)
```

```python
font_type = 'Times New Roman'
xlabel_size = 15
xlabel_pad = 5

ylabel_size = 15
ylabel_pad = 10

title_size = 20
title_pad = 20

font = font_manager.FontProperties(family='Times New Roman',
                                   style='normal')


plt.figure()

scatter1 = plt.scatter(sizes, accs, marker = 'x', s = 100, label = 'Seed Accuracies')
scatter2 = plt.scatter(sizes[0:14:5], avg_accs, s = 100, label = 'Mean Accuracy')
plt.ylim(0.88,1)
plt.ylabel('Accuracy', labelpad = ylabel_pad, fontsize = ylabel_size, fontfamily = font_type)
plt.xlabel('Community Size', labelpad = xlabel_pad, fontsize = xlabel_size, fontfamily = font_type)
#plt.legend((scatter1, scatter2), ('Seed Accuracies', 'Mean Accuracy'), fontfamily = font_type)
plt.legend(prop=font)
plt.grid(True, linewidth = 0.7)
plt.xlim(-0.5,2.5)
plt.title('Game Success', fontsize = title_size, pad = title_pad, fontfamily = font_type)

plt.show()
```
