```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager

plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

```python
comm_sizes = [1,2,4]

sizes = []
scores = []
avg_scores = []

with open('/home/jmorisse/Desktop/BA Thesis/Bachelor-Thesis/Code/Data/Results/concept_encoding_scores.txt', 'r') as f:
    for i,line in enumerate(f):
        seed_scores = [float(score) for score in line.split(';')[0].split(',')]
        for j in range(len(seed_scores)):
            sizes.append(str(comm_sizes[i]))
        seed_avg = float(line.split(';')[1])
        
        for seed_score in seed_scores:
            scores.append(seed_score)
        avg_scores.append(seed_avg)

print(sizes)
print(scores)
print(avg_scores)
```

```python
font_type = 'Times New Roman'
xlabel_size = 15
xlabel_pad = 5

ylabel_size = 15
ylabel_pad = 10

title_size = 20
title_pad = 20

font = font_manager.FontProperties(family='Times New Roman',
                                   style='normal')


plt.figure()

scatter1 = plt.scatter(sizes, scores, marker = 'x', s = 100, label = 'Seed Score')
scatter2 = plt.scatter(sizes[0:14:5], avg_scores, s = 100, label = 'Mean Score')
plt.ylim(0,0.01)
plt.ylabel('Accuracy of Diagn. Classifier', labelpad = ylabel_pad, fontsize = ylabel_size, fontfamily = font_type)
plt.xlabel('Community Size', labelpad = xlabel_pad, fontsize = xlabel_size, fontfamily = font_type)
#plt.legend((scatter1, scatter2), ('Seed Accuracies', 'Mean Accuracy'), fontfamily = font_type)
plt.legend(prop=font)
plt.grid(True, linewidth = 0.7)
plt.xlim(-0.5,2.5)
plt.title('Concept Encoding', fontsize = title_size, pad = title_pad, fontfamily = font_type)

plt.show()
```
