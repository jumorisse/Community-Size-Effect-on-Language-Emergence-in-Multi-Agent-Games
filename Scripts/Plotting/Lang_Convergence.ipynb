```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager

plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

```python
comm_sizes = [1,2,4]

sizes = []
conv_scores = []
avg_conv_scores = []

with open('/home/jmorisse/Desktop/BA Thesis/Bachelor-Thesis/Code/Data/Results/conv_scores.txt', 'r') as f:
    for i,line in enumerate(f):
        seed_conv_scores = [float(conv_score) for conv_score in line.split(';')[0].split(',')]
        for j in range(len(seed_conv_scores)):
            sizes.append(str(comm_sizes[i]))
        seed_avg = float(line.split(';')[1])
        
        for seed_conv_score in seed_conv_scores:
            conv_scores.append(seed_conv_score)
        avg_conv_scores.append(seed_avg)

print(sizes)
print(conv_scores)
print(avg_conv_scores)
```

```python
font_type = 'Times New Roman'
xlabel_size = 15
xlabel_pad = 5

ylabel_size = 15
ylabel_pad = 10

title_size = 20
title_pad = 20

font = font_manager.FontProperties(family='Times New Roman',
                                   style='normal')


plt.figure()

scatter1 = plt.scatter(sizes, conv_scores, marker = 'x', s = 100, label = 'Seed Scores')
scatter2 = plt.scatter(sizes[0:14:5], avg_conv_scores, s = 100, label = 'Mean Scores')
plt.ylim(0,1.05)
plt.ylabel('1 - Hamming Distance', labelpad = ylabel_pad, fontsize = ylabel_size, fontfamily = font_type)
plt.xlabel('Community Size', labelpad = xlabel_pad, fontsize = xlabel_size, fontfamily = font_type)
#plt.legend((scatter1, scatter2), ('Seed Accuracies', 'Mean Accuracy'), fontfamily = font_type)
plt.legend(prop=font)
plt.grid(True, linewidth = 0.7)
plt.xlim(-0.5,2.5)
plt.title('Language Convergence', fontsize = title_size, pad = title_pad, fontfamily = font_type)

plt.show()
```
