```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager

plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
```

```python
comm_sizes = [1,2,4]

sizes = []
top_sims = []
avg_top_sims = []

with open('/home/jmorisse/Desktop/BA Thesis/Bachelor-Thesis/Code/Data/Results/comm_top_sims.txt', 'r') as f:
    for i,line in enumerate(f):
        seed_top_sims = [float(top_sim) for top_sim in line.split(';')[0].split(',')]
        for j in range(len(seed_top_sims)):
            sizes.append(str(comm_sizes[i]))
        seed_avg = float(line.split(';')[1])
        
        for seed_top_sim in seed_top_sims:
            top_sims.append(seed_top_sim)
        avg_top_sims.append(seed_avg)

print(sizes)
print(top_sims)
print(avg_top_sims)
```

```python
font_type = 'Times New Roman'
xlabel_size = 15
xlabel_pad = 5

ylabel_size = 15
ylabel_pad = 10

title_size = 20
title_pad = 20

font = font_manager.FontProperties(family='Times New Roman',
                                   style='normal')


plt.figure()

scatter1 = plt.scatter(sizes, top_sims, marker = 'x', s = 100, label = 'Seed Top. Sim.')
scatter2 = plt.scatter(sizes[0:14:5], avg_top_sims, s = 100, label = 'Mean Top. Sim.')
plt.ylim(0.15,0.35)
plt.ylabel('Topographic Similarity', labelpad = ylabel_pad, fontsize = ylabel_size, fontfamily = font_type)
plt.xlabel('Community Size', labelpad = xlabel_pad, fontsize = xlabel_size, fontfamily = font_type)
#plt.legend((scatter1, scatter2), ('Seed Accuracies', 'Mean Accuracy'), fontfamily = font_type)
plt.legend(prop=font)
plt.grid(True, linewidth = 0.7)
plt.xlim(-0.5,2.5)
plt.title('Topographic Similarity', fontsize = title_size, pad = title_pad, fontfamily = font_type)

plt.show()
```
